```python
# Take integer inputs from user until he/she presses q ( Ask to press q to quit after every integer input ).
# Print average and product of all numbers.

numbers = []
while True:
    s = input("Enter a number (or 'q' to quit): ")
    if s == 'q':
        break
    numbers.append(int(s))

if numbers:
    avg = sum(numbers) / len(numbers)
    prod = 1
    for n in numbers:
        prod *= n
    print("Average:", avg)
    print("Product:", prod)
else:
    print("No numbers entered.")
```

```python
# Write a program to find greatest common divisor (GCD) or highest common factor (HCF) of given two numbers.
# Take two numbers as input

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
x, y = a, b
while y != 0:
    x, y = y, x % y
print(f"GCD/HCF of {a} and {b} is {x}")
```

```python
# A three digit number is called Armstrong number if sum of cube of its digit is equal to number itself.
#E.g.- 153 is an Armstrong number because (13)+(53)+(33) = 153.
# Write all Armstrong numbers between 100 to 500.

for num in range(100, 501):
  hundreds=num//100
  tens=(num//10) % 10
  units=num%10
  sum_cubes=hundreds**3+tens**3+units**3
  if num==sum_cubes:
    print(num)
```

```python
# Write a program to find prime factor of a number.
# If a factor of a number is prime number then it is its prime factor.

n=int(input("Enter a Number: "))
factor=2

while factor<=n:
  if n%factor==0:
    print(factor)
    n=n/factor
  else:
    factor+=1
```

```python
# Write a program to print all prime number in between 1 to 100
# Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
for num in range(1, 101):
    if is_prime(num):
        print(num, end=' ')
```
