```python
4# Take 10 integer inputs from user and store them in a list. Again ask user to give a number.
# Now, tell user whether that number is present in list or not.
# hint( Iterate over list using while loop ).

nums=[]
count=0
while count<10:
  value=int(input("Enter a Number: "))
  nums.append(value)
  count+=1

check=int(input("Enter a number to check: "))

if check in nums:
  print(check,"is present in list")
else:
  print(check,"is Not present in list")

```

```python
# Take 10 integer inputs from user and store them in a list. Now, copy all the elements in another list but in reverse order.
lst=[]
cnt=0
while cnt<10:
  value=int(input("Enter a Number: "))
  lst.append(value)
  cnt+=1

reversed_lst = lst[::-1]
print("Original list:", lst)
print("Reversed list:", reversed_lst)
```

```python
# Input two 3x3 matrices
A = [[int(input(f"A[{i}][{j}]: ")) for j in range(3)] for i in range(3)]
B = [[int(input(f"B[{i}][{j}]: ")) for j in range(3)] for i in range(3)]

# Matrix addition
add = [[A[i][j] + B[i][j] for j in range(3)] for i in range(3)]

# Matrix multiplication
mult = [[sum(A[i][k] * B[k][j] for k in range(3)) for j in range(3)] for i in range(3)]

print("\nSum of matrices:")
for row in add:
    print(row)
print("\nProduct of matrices:")
for row in mult:
    print(row)

```
