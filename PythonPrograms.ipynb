```python
#2.Check if a year is a leap year. A leap year is a year that is divisible by 4, but if it is a
# century year (divisible by 100), it must also be divisible by 400 to be a leap year

n=int(input("Enter a year: "))
if (n%4==0 and n%100!=0) or (n%400==0):
  print(n,"is a leap year")
else:
  print(n,"is not a leap year")
```

```python
#4.Write a Python program to assign the grades to each subject based on the grade scale:
#Student subject marks:
#maths = 85
#science = 78
#english = 92
#Grade scale:
#A: 90-100
#B: 80-89
#C: 70-79
#D: 60-69
#E: 50-59
#F: Below 50

maths = float(input("Enter Maths Marks: "))
science = float(input("Enter Science Marks: "))
english = float(input("Enter English Marks: "))

def grade(m):
    if m >= 90:
        return 'A'
    elif m >= 80:
        return 'B'
    elif m >= 70:
        return 'C'
    elif m >= 60:
        return 'D'
    elif m >= 50:
        return 'E'
    else:
        return 'F'

print("Maths:", grade(maths))
print("Science:", grade(science))
print("English:", grade(english))
```

Mathematical Programs


```python
#1.Check if a number is prime or not.
n = int(input("Enter a number: "))
if n>1:
  for i in range(2,n):
    if n%i==0:
      print(n,"is not a prime number")
      break
  else:
    print(n,"is a prime number")
```

```python
#2.Find the sum of digits of a number (eg: input: 1234 and output: 10)
num=int(input("Enter a number: "))
sum_of_digits = 0
while num > 0:
  digit = num % 10
  sum_of_digits += digit
  num //= 10
print("Sum of digits:", sum_of_digits)
```

```python
#3.Find the LCM of two numbers
a, b = 12, 18
def lcm(x, y):
  greater = max(x, y)
  while True:
      if greater % x == 0 and greater % y == 0:
          return greater
      greater += 1
print("LCM:", lcm(a, b))
```

```python
#4.Find the GCD of two numbers
import math
print("GCD:", math.gcd(a, b))
```

```python
#4.Lambda function inside map()
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print("Squared numbers using map():", squared_numbers)
```
