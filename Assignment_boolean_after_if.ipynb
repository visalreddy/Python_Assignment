```python
# Take two inputs from user and check whether they are equal or not.
n1=int(input("Enter first number:"))
n2=int(input("Enter second number:"))
if n1==n2:
  print("Numbers are equal")
else:
  print("Numbers are not equal")
```

```python
# Take 3 inputs from user and check :
#all are equal
#any of two are equal

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
if a==b==c:
  print("All numbers are equal.")
elif a==b or b==c or a==c:
  print("Two numbers are equal.")
else:
  print("All numbers are distinct.")
```

```python
# Take two number and check whether the sum is greater than 5, less than 5 or equal to 5.
num1=int(input("Enter number 1: "))
num2=int(input("Enter number 2: "))

sum=num1+num2
if sum>5:
  print("Sum is greater than 5.")
elif sum<5:
  print("Sum is less than 5.")
else:
  print("Sum is equal to 5.")
```

```python
# Suppose passing marks of a subject is 35.
# Take input of marks from user and check whether it is greater than passing marks or not.

marks=int(input("Enter marks: "))
if marks>35:
  print("Marks are greater than passing marks.")
elif marks<35:
  print("Marks are less than passing marks.")
else:
  print("Marks are equal to passing marks.")
```
