```python
#86.Create a customer lookup DataFrame with unique CustomerID, CustomerName, and total orders. Save it.
sales = pd.read_csv("Advanced_Dirty_Sales_Dataset.csv")

customer_lookup = (
    sales.groupby(['CustomerID', 'CustomerName'])['OrderID']
         .nunique()
         .reset_index()
         .rename(columns={'OrderID': 'TotalCustomerOrders'}))

customer_lookup.to_csv("customer_lookup.csv", index=False)
print(customer_lookup.head())
```

```python
#87.Create a product lookup DataFrame with unique Product, Category, and average Price. Save it.
sales = pd.read_csv("Advanced_Dirty_Sales_Dataset.csv")
sales['Price'] = sales['Price'].astype(str).str.replace('$', '', regex=False).astype(float)

product_lookup = (
    sales.groupby(['Product', 'Category'])['Price']
         .mean()
         .reset_index()
         .rename(columns={'Price': 'AvgProductPrice'}))

product_lookup.to_csv("product_lookup.csv", index=False)
```

```python
#88. Merge the original dataset with both lookups.
merged_1 = sales.merge(customer_lookup, on=['CustomerID', 'CustomerName'], how='left')
merged_2 = merged_1.merge(product_lookup, on=['Product', 'Category'], how='left')
```

```python
#89. After merge, create a master file with these columns: OrderID, OrderDate,CustomerName, Product, Category, Price, Quantity, TotalAmount, Returned, and theAvgProductPrice and TotalCustomerOrders from lookup tables.
master_file = merged_2[
    [
        'OrderID',
        'OrderDate',
        'CustomerName',
        'Product',
        'Category',
        'Price',
        'Quantity',
        'TotalAmount',
        'Returned',
        'AvgProductPrice',
        'TotalCustomerOrders'
    ]
]

master_file.to_csv("master_sales_file.csv", index=False)
print(master_file)
```

```python
#90. Analyze and visualize the top 5 customers and top 5 products contributing to revenue using the master file.
master_file['TotalAmount'] = pd.to_numeric(master_file['TotalAmount'], errors='coerce')

top_customers = (master_file.groupby('CustomerName')['TotalAmount'].sum().sort_values(ascending=False).head(5))
print("\nTop 5 Customers by Revenue:")
print(top_customers)

top_products = (master_file.groupby('Product')['TotalAmount'].sum().sort_values(ascending=False).head(5))
print("\nTop 5 Customers by Revenue:")
print(top_products)
```
