```python
#88.How do you compute inner, outer, sum, and multiplication using einsum?
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a)
print(b)

inner = np.einsum('i,i->', a, b)
print("\nInner: ",inner)

outer = np.einsum('i,j->ij', a, b)
print("\nOuter:\n", outer)

total_sum = np.einsum('i->', a)
print("\nSum: ", total_sum)

elementwise = np.einsum('i,i->i', a, b)
print("\nMultiplication: ", elementwise)
```
